```python
%matplotlib inline 
import matplotlib.pylab
import numpy as np
import pandas as pd
```

```python
demand_data = pd.read_csv("./Demand_for_Texas_(region)_hourly_-_UTC_time.csv", skiprows=4, header = 0, parse_dates = [0], names = ['Time', 'MWhs'], index_col = 0)
```

## Convert UTC To CST Time

```python
demand_data.index = demand_data.index.tz_localize('UTC')
demand_data.index = demand_data.index.tz_convert('US/Central')
```

# Yearly Data 2020

```python
demand_data20 = demand_data['2020']
```

```python
demand_data20.plot()
```

#    Monthly Data

```python
demand_may2020 = demand_data20[demand_data20.index.month == 5]
```

```python
demand_may2020.plot()
```

# Weekly Data May25 - May 31

```python
demand_may_week = demand_may2020[demand_may2020.index.week == 22]
```

```python
demand_may_week.plot()
```

# Daily Data: May 31 (Sunday)

```python
demand_may31 = demand_may_week[demand_may_week.index.day == 31]
```

```python
demand_may31.plot()
```
